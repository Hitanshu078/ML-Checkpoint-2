```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def perform_eda_for_outliers(df):
    """
    Generates Boxplots and Histograms to distinguish between
    natural outliers and data entry errors (noise).
    """
    # These are the columns we suspect contain "placeholder" or "impossible" values
    suspicious_cols = [
        'triglycerides',
        'ggt_enzyme_level',
        'vision_left',
        'vision_right',
        'hdl_cholesterol',
        'ldl_cholesterol'
    ]

    # Filter strictly for columns that actually exist in your dataframe
    cols_to_plot = [c for c in suspicious_cols if c in df.columns]

    # --- PART 1: VISUALIZATION (Boxplots & Histograms) ---
    n_cols = 2
    n_rows = (len(cols_to_plot) + 1) // 2

    plt.figure(figsize=(15, 5 * n_rows))
    
    for i, col in enumerate(cols_to_plot):
        plt.subplot(n_rows, n_cols, i+1)
        
        # We use a boxplot because it clearly shows the "Gap" between 
        # normal data and the "999" error codes.
        sns.boxplot(x=df[col], color='skyblue')
        plt.title(f'Boxplot of {col}')
        plt.grid(True, alpha=0.3)
        
    plt.tight_layout()
    plt.show()

    # --- PART 2: STATISTICAL CONFIRMATION ---
    print("\n" + "="*40)
    print("SPECIFIC NOISE CHECKS")
    print("="*40)

    # Check for the specific "999" placeholder code
    for col in ['triglycerides', 'ggt_enzyme_level']:
        if col in df.columns:
            count = len(df[df[col] == 999])
            print(f"Column '{col}':")
            print(f"  -> Rows with value 999 (Noise): {count}")
            print(f"  -> Min value: {df[col].min()}")
            print(f"  -> Max value: {df[col].max()}")
            print("-" * 20)

    # Check for the "9.9" vision code
    for col in ['vision_left', 'vision_right']:
        if col in df.columns:
            count = len(df[df[col] == 9.9])
            print(f"Column '{col}':")
            print(f"  -> Rows with value 9.9 (Blind/Missing): {count}")
            print("-" * 20)

    # Check for Impossible Cholesterol Values
    if 'hdl_cholesterol' in df.columns:
        # Anything above 200 is extremely suspicious for HDL
        count_high = len(df[df['hdl_cholesterol'] > 200])
        print(f"Column 'hdl_cholesterol':")
        print(f"  -> Rows > 200 (Physiologically Impossible): {count_high}")
        
    if 'ldl_cholesterol' in df.columns:
        # Anything above 600 is extremely suspicious for LDL
        count_high = len(df[df['ldl_cholesterol'] > 600])
        print(f"Column 'ldl_cholesterol':")
        print(f"  -> Rows > 600 (Likely Error): {count_high}")

    print("="*40)

# --- EXECUTE ---
# Assuming your dataframe is named 'df'
perform_eda_for_outliers(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

def process_datasets():
    # 1. Load Datasets
    train_df = pd.read_csv('../data_files/train.csv')
    test_df = pd.read_csv('../data_files/test.csv')
    
    print(f"Original Train shape: {train_df.shape}")
    print(f"Original Test shape:  {test_df.shape}")

    # 2. Separate Target (y) and Features (X)
    # We define the target column name
    target_col = 'has_copd_risk'
    
    # Create y_train
    if target_col in train_df.columns:
        y_train = train_df[target_col]
        # Drop target from X_train
        X_train = train_df.drop(columns=[target_col])
    else:
        raise ValueError(f"Target column '{target_col}' not found in training data!")

    # Prepare X_test (Ensure target is not in it, just in case)
    if target_col in test_df.columns:
        X_test = test_df.drop(columns=[target_col])
    else:
        X_test = test_df.copy()

    # 3. Drop Useless Columns (ID and Zero Variance)
    # Apply to BOTH X_train and X_test
    cols_to_drop = ['patient_id', 'oral_health_status']
    
    X_train = X_train.drop(columns=[c for c in cols_to_drop if c in X_train.columns])
    X_test = X_test.drop(columns=[c for c in cols_to_drop if c in X_test.columns])

    # 4. Encode Categorical Features (Manual Mapping)
    # We map them ensuring consistency across both datasets
    for df in [X_train, X_test]:
        # Sex: M -> 0, F -> 1 (or vice versa, just needs to be consistent)
        if 'sex' in df.columns:
            df['sex'] = df['sex'].map({'M': 0, 'F': 1})
        
        # Tartar: N -> 0, Y -> 1
        if 'tartar_presence' in df.columns:
            df['tartar_presence'] = df['tartar_presence'].map({'N': 0, 'Y': 1})

    # 5. CLEANING: Handle specific noise/outliers (Function applied to both)
    def clean_hard_errors(df):
        df_clean = df.copy()
        
        # Replace 999 with NaN (Triglycerides, GGT)
        for col in ['triglycerides', 'ggt_enzyme_level']:
            if col in df_clean.columns:
                df_clean.loc[df_clean[col] == 999, col] = np.nan
        
        # Replace 9.9 with NaN (Vision)
        for col in ['vision_left', 'vision_right']:
            if col in df_clean.columns:
                df_clean.loc[df_clean[col] == 9.9, col] = np.nan

        # Replace Impossible Biological Values
        if 'hdl_cholesterol' in df_clean.columns:
            df_clean.loc[df_clean['hdl_cholesterol'] > 200, 'hdl_cholesterol'] = np.nan
        
        if 'ldl_cholesterol' in df_clean.columns:
            df_clean.loc[df_clean['ldl_cholesterol'] > 600, 'ldl_cholesterol'] = np.nan
            
        return df_clean

    X_train = clean_hard_errors(X_train)
    X_test = clean_hard_errors(X_test)

    # 6. IMPUTATION (Fit on Train -> Transform Test)
    # Strategy: Vision uses MEAN, Biological/Others use MEDIAN
    vision_cols = [c for c in ['vision_left', 'vision_right'] if c in X_train.columns]
    other_cols = [c for c in X_train.columns if c not in vision_cols]

    # Impute Vision (Mean)
    if vision_cols:
        imputer_mean = SimpleImputer(strategy='mean')
        # Learn mean from Train
        X_train[vision_cols] = imputer_mean.fit_transform(X_train[vision_cols])
        # Apply Train's mean to Test
        X_test[vision_cols] = imputer_mean.transform(X_test[vision_cols])

    # Impute Others (Median)
    if other_cols:
        imputer_median = SimpleImputer(strategy='median')
        # Learn median from Train
        X_train[other_cols] = imputer_median.fit_transform(X_train[other_cols])
        # Apply Train's median to Test
        X_test[other_cols] = imputer_median.transform(X_test[other_cols])

    # 7. SCALING (Standardization)
    # CRITICAL: Fit on Train, Transform Test
    scaler = StandardScaler()
    
    # Fit & Transform Train
    X_train_scaled_array = scaler.fit_transform(X_train)
    
    # Only Transform Test (using Train's mean/std)
    X_test_scaled_array = scaler.transform(X_test)

    # Convert back to DataFrames for convenience
    X_train_final = pd.DataFrame(X_train_scaled_array, columns=X_train.columns)
    X_test_final = pd.DataFrame(X_test_scaled_array, columns=X_test.columns)

    print("\n--- Processing Complete ---")
    print(f"Final X_train shape: {X_train_final.shape}")
    print(f"Final X_test shape:  {X_test_final.shape}")
    
    return X_train_final, y_train, X_test_final

# --- EXECUTE ---
# Replace with your actual file names
X_train_scaled, y_train, X_test_scaled = process_datasets()

# Check the first few rows
print(X_train_scaled.head())
```
